#GroupBy and Aggregrate


````python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Flavors.csv')
df
````

````python
group_by_frame = df.groupby('Base Flavor')
# print(group_by_frame.sum())
print(group_by_frame.count())
print(group_by_frame.mean(numeric_only=True))
print(group_by_frame.describe())

````

````python
df.groupby(['Base Flavor','Liked']).agg({'Flavor Rating':['mean','max','count','sum']})
````

````python
df.groupby('Base Flavor').describe()
````

# Merge, join and concatenate

````python
df1=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/LOTR.csv')
df1
````

````python
df2=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/LOTR 2.csv')
df2
````

````python
pd.concat([df1,df2],ignore_index=True)
````

innerjoin intersection
outerjoin union
leftjoin A-B only A
rightjoin B-A only B

````python
#inner join
df1.merge(df2,how='inner',on=['FellowshipID','FirstName'])
````

````python
df1.merge(df2,how='outer')
````

````python
df1.merge(df2,how='left',on=['FellowshipID','FirstName'])
````

````python
df1.merge(df2,how='right',on=['FellowshipID','FirstName'])
````
