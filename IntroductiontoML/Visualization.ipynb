# Matpilotlib and seaborn


Why choose matpilotlib and seaborn

Both libaries are among the most popular choies for data visualization in Python. Here's why: Matpilotlib : know for its flexibility, Matpilot allows detailed customization of visual elements, making it perfect for tailored plots.
Seaborn: Build on top of Matpilotlib, Seaborn simplifies visualization tasks by offering tasks by offering high-level functions to create aesthically pleasing statistical graphical.


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, label ="Sin wave")
plt.title("Line plot example")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
# map ko harek le k janauxx legend
plt.legend()
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D','E']
values = [10, 20, 30, 40,90]

plt.bar(categories, values,color=['red','green','yellow','purple','blue'])
plt.title("Bar Chart Example")
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
import seaborn as sns
```

```python
#Histogram and KDE plot:
# rand gives random values of range 0 to 1 and randn gives random values of range -infinite to infinite
data = np.random.randn(1000)
sns.histplot(data, kde=True,color='green')
plt.show()
```

```python
#heat map
matrix = np.random.rand(5,5)
matrix
```

```python
sns.heatmap(matrix,annot=True,cmap='RdPu')
plt.title("Heatmap Example")
plt.show()
```

```python
sns.set_style("whitegrid")
df= sns.load_dataset('iris')
print(df)
# //cirle-o square-s diamond-D
sns.pairplot(df,hue='species',markers=["o","s","D"])
plt.show()
```
